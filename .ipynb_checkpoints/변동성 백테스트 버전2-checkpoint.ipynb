```python
import os
import pickle

import logging
import pickle

logging.basicConfig(filename="backtest1.txt", level=logging.INFO)


def load_data(file_name):
    try:
        f = open("./data/{}".format(file_name), "rb")
        data = pickle.load(f)
        f.close()
        return data
    except:
        pass
```

```python
def process_df(df):
    
    df = df[df['volume'] !=0]
    ma5 = df['close'].rolling(window=5).mean()
    ma20 = df['close'].rolling(window=20).mean()
    ma60 = df['close'].rolling(window=60).mean()
    df.insert(len(df.columns), "MA5", ma5)
    df.insert(len(df.columns), "MA20", ma20)
    df.insert(len(df.columns), "MA60", ma60)
    df.dropna(inplace=True)
    index_list = df.index
    return df
```

```python
def test(df):
    initial = 0
    index_list = df.index
    cnt = 0
    for i in range(1, len(df)-1):
        
        
        diff = df.loc[index_list[i-1], 'high'] - df.loc[index_list[i-1], 'low']
        if df.loc[index_list[i],'close'] > df.loc[index_list[i],'open']+diff :
            if df.loc[index_list[i-1], 'MA5'] > df.loc[index_list[i-1], 'MA20'] and  df.loc[index_list[i-1], 'MA60'] > df.loc[index_list[i-1], 'MA5']:
            

                initial = initial - df.loc[index_list[i],'close']
                initial = initial + df.loc[index_list[i+1],'open']
                cnt = cnt+1

    logging.info("시가{} 손익{} 카운터{}".format(df.loc[index_list[1], 'open'], initial, cnt))
```

```python
for i in os.listdir("data"):
    df = load_data(i)
    df = process_df(df)
    logging.info("{}".format(i))
    test(df)
    print("{} done".format(i))
```
